```python
import json

# Define the data description dictionary
data_description = {
    "Country": {
        "type": "string",
        "description": "Name of the country for which the data is collected"
    },
    "Year": {
        "type": "integer",
        "description": "Calendar year from 2001 to 2020"
    },
    "Primary Enrollment": {
        "type": "float",
        "description": "Total number of students enrolled in primary education for that year"
    },
    "GDP per Capita": {
        "type": "float",
        "description": "Gross Domestic Product per capita (current US$)"
    }
}

# Save it as a JSON file
with open("data_description.json", "w") as f:
    json.dump(data_description, f, indent=4)

print("data_description.json file has been created!")
```

```python
# Read and display the JSON file
with open("data_description.json", "r") as f:
    data = json.load(f)

data
```
